##### **Problem 03:**
Find the mean of the 3rd column of the matrix without using the built-in function and loop.

```python
# Write your code here
third_column = matrix[:, 2]

# Calculate the mean without using built-in functions or loops
# Mean = sum of elements / number of elements
total_sum = 0
num_elements = 0

for value in third_column:
    total_sum += value
    num_elements += 1

mean_value = total_sum / num_elements

print(f"Mean of the 3rd column: {mean_value}")
```

##### **Problem 05:**
Write a function that takes a number and returns a vector like the previous problem. For example, if the input is 5, then the output should be `[25, 16, 9, 4, 1]`. If the input is 10, then the output should be `[100, 81, 64, 49, 36, 25, 16, 9, 4, 1]`.

```python
import numpy as np

# Write your code here
a, b = int(input("Enter a: ")), int(input("Enter b: "))
def generate_squared_vector(n):
    # Generate a vector of squares from n down to 1
    vector = np.array([i**2 for i in range(n, 0, -1)])
    return vector

# Test the function
print("Output for input 5:", generate_squared_vector(a))
print("Output for input 10:", generate_squared_vector(b))
```

##### **Problem 07:**
Plot $y = 2x^2 + 3x + 1$ for $x \in [-3, 3]$ with interval 0.25. You can use any built-in function to plot the graph.

```python
# Write your code here
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 2*x**2 + 3*x + 1

# Create an array of x values from -3 to 3 with an interval of 0.25
x_values = np.arange(-3, 3.25, 0.25)
y_values = f(x_values)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = 2x + 3x + 1', color='blue')
plt.title('Plot of y = 2x + 3x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
# Write your code here
from mpl_toolkits.mplot3d import Axes3D

# Define the ranges for x and y
x_range = np.arange(-3, 3.5, 0.5)
y_range = np.arange(-3, 3.5, 0.5)

# Create a meshgrid for x and y
X, Y = np.meshgrid(x_range, y_range)

# Calculate Z based on the equation x + y + z = 1
Z = 1 - X - Y

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Surface Plot of x + y + z = 1')

# Show the plot
plt.show()
```

Download the image of Lena Forsen from the following link: [https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png](https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png). Then, do the following tasks:

```python
# Write your code here
import imageio
from skimage.color import rgb2gray
from skimage.transform import resize

# Step 1: Load the image to a numpy array
url = 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png'
image = imageio.imread(url)

# Step 2: Find the size of the image
image_size = image.shape
print(f"Image size: {image_size}")

# Step 3: Find the minimum and maximum value of the image
min_value = np.min(image)
max_value = np.max(image)
print(f"Minimum value: {min_value}, Maximum value: {max_value}")

# Step 4: Find the mean and standard deviation of the image
mean_value = np.mean(image)
std_value = np.std(image)
print(f"Mean value: {mean_value}, Standard deviation: {std_value}")

# Step 5: Plot the image
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')
plt.show()

# Step 6: Resize the image to 50 x 50
resized_image = resize(image, (50, 50), anti_aliasing=True)

# Step 7: Convert the image to grayscale
gray_image = rgb2gray(resized_image)

# Step 8: Plot the grayscale image
plt.figure(figsize=(6, 6))
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title('Grayscale Image')
plt.show()

# Step 9: Save the grayscale image
imageio.imwrite('grayscale_image.png', (gray_image * 255).astype(np.uint8))

# Step 10: Stack the columns of the image to form a vector of size 2500 x 1
image_vector = gray_image.flatten().reshape(-1, 1)
print(f"Image vector shape: {image_vector.shape}")
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
# Write your code here
x_values = np.arange(-1, 1.01, 0.01)

# Define the true functions
true_e2x = np.exp(2 * x_values)
true_e3x = np.exp(3 * x_values)
true_e5x = np.exp(5 * x_values)

# Compute 2nd order Taylor approximations around a = 0
def taylor_e2x(x):
    return 1 + 2 * x + (2 * x)**2 / 2

def taylor_e3x(x):
    return 1 + 3 * x + (3 * x)**2 / 2

def taylor_e5x(x):
    return 1 + 5 * x + (5 * x)**2 / 2

# Compute Taylor approximations
taylor_2x = taylor_e2x(x_values)
taylor_3x = taylor_e3x(x_values)
taylor_5x = taylor_e5x(x_values)

# Method 1: Taylor approximation separately for each e2x and e3x, then multiply
method_1_approx = taylor_2x * taylor_3x

# Method 2: Direct Taylor approximation for e5x
method_2_approx = taylor_5x

# Calculate Mean Square Error (MSE) for both methods
mse_method_1 = np.mean((true_e5x - method_1_approx)**2)
mse_method_2 = np.mean((true_e5x - method_2_approx)**2)

print(f'MSE Method 1: {mse_method_1}')
print(f'MSE Method 2: {mse_method_2}')

# Plotting the results
plt.figure(figsize=(10, 6))

# Original functions
plt.plot(x_values, true_e5x, label='e^(5x)', color='black', linewidth=2)

# Method 1 approximation
plt.plot(x_values, method_1_approx, label='Method 1: Taylor(e^(2x)) * Taylor(e^(3x))', linestyle='dashed')

# Method 2 approximation
plt.plot(x_values, method_2_approx, label='Method 2: Taylor(e^(5x))', linestyle='dotted')

# Set titles and labels
plt.title('Taylor Approximation of e^(5x)')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.legend()
plt.grid()
plt.show()
```
